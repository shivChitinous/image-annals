# Stacks and Channels

```python
import matplotlib.pyplot as plt
import imageio as img
import numpy as np
import scipy as sp
import skimage as ski #for different image analysis algorithms
import skimage.filters
import astropy.nddata as ap
import seaborn as sns; sns.set_style('white'); sns.set_context('notebook')
```

```python
pic = np.array(img.imread('Images/fetter_Celegans_cellfig10.jpg')).astype('double')
plt.figure(figsize = (5,5))
plt.imshow(pic, cmap = 'gray');
```

```python
print(np.shape(pic))
```

2D grayscale images are represented as (*row*, *col*) - *row* is number of rows and *col* is the number of columns.

```python
pic2 = np.array(img.imread('Images/k_icebird_circles_28Mar2014.png'))
plt.figure(figsize = (5,5))
plt.imshow(pic2, cmap = 'gray');
```

```python
print(np.shape(pic2))
```

Coloured imaages are represented as (*row*, *col*, *ch*) - *ch* is a channel (*R*, *G*, *B* and *A*)

## Converting coloured image to grayscale

```python
pic2_gr = ski.color.rgb2gray(ski.color.rgba2rgb(pic2))
```

```python
np.shape(pic2_gr)
```

## Stacking grayscale images

Stacking grayscale images _N_ times.

```python
N = 10
```

```python
stack = np.zeros((np.shape(pic2_gr)[0],np.shape(pic2_gr)[1],N))
for i in range(np.shape(stack)[2]):
    stack[:,:,i] = pic2_gr
```

```python
np.shape(stack)
```

```python
plt.figure(figsize = (5,5))
plt.imshow(stack[:,:,1], cmap = 'gray');
```
